## 1) Does discount amount have a statistically significant effect on the quantity of a product in an order? If so, at what levels of discount?

<b>Null hypothesis</b>: The average order quantities between orders with no discount and a discount are the same.

<b>Alternative hypothesis</b>: The average order quantities between orders with no discount and a discount are different. 

<b>Alpha level = </b> <b>0.05</b>

```python
conn = sqlite3.connect("Northwind_small.sqlite")
cur = conn.cursor()
```

```python
cur.execute("""select name from sqlite_master where type= 'table'""").fetchall()
```

```python
cur.execute("""SELECT Quantity, Discount
               FROM OrderDetail;""")
df_quan_disc = pd.DataFrame(cur.fetchall())
df_quan_disc.columns = [x[0] for x in cur.description]
df_quan_disc.head(12)
```

```python
# Query above gives us the quantity ordered and what level of discount(if any) was applied, now we check all the discount levels.
df_quan_disc["Discount"].value_counts()
```

```python
# Sneak peak into how discounts could influence the amount of products per order.   
sns.set(rc = {"figure.figsize" : (8, 8)})
sns.set(color_codes = True)
custom_palette = sns.color_palette("Paired", 11)
sns.barplot(df_quan_disc["Discount"], df_quan_disc["Quantity"], ci = False, palette = custom_palette)
plt.title("Quantity of Products Ordered at Different Discount Levels")
plt.show();
```

Looking at this plot we can definitely see that there are particular discount levels that might influence the amount of products ordered.

```python
# Now I will separate the products with a discount from the ones with no discount.
df_dsc = df_quan_disc[df_quan_disc["Discount"] > 0].reset_index(drop = True)
df_dsc.head()
```

```python
df_no_dsc = df_quan_disc[df_quan_disc["Discount"] == 0].reset_index(drop = True)
df_no_dsc.head()
```

```python
# Let's check for normality
sns.set_style("darkgrid")
sns.distplot(df_dsc["Quantity"]);
```

```python
sns.distplot(df_no_dsc["Quantity"]);
```

##### It seems that our distributions are positively skewed and are not as normal as we would like. I will put the Central Limit Theorem to work here.

```python
samp_no_dsc = []
for i in range(150):
    samples = df_no_dsc["Quantity"].sample(n = 70)
    samp_mean = np.mean(samples)
    samp_no_dsc.append(samp_mean)
sns.distplot(samp_no_dsc);
```

```python
samp_dsc = []
for j in range(150):
    samples2 = df_dsc["Quantity"].sample(n = 70)
    samp_mean2 = np.mean(samples2)
    samp_dsc.append(samp_mean2)
sns.distplot(samp_dsc);
```

```python
# Now that the distributions are a bit more normal lets have a quick look at their differences.
sns.distplot(samp_no_dsc, label = "No Discount", kde = False)
sns.distplot(samp_dsc, label = "With Discount", kde = False)
plt.legend()
plt.xlabel("Product Quantity")
plt.ylabel("Frequency")
plt.title("Average Order Quantities")
plt.show()
```

Here we can see that there seems to be more orders with higher quantities when a discount is applied. Lets run a Two Sample T-test to verify the observation.

```python
results = stats.ttest_ind(samp_no_dsc, samp_dsc, equal_var= False)
print("The p-value: ", results[1])
if results[1] < 0.05:
    print("Null Hypothesis rejected due to a p-value lower than our alpha of 0.05")
else:
    print("Failure to reject Null Hypothesis")
```

The Welch's T-test confirms our observation with a p-value well below our <b>alpha</b> of <b>0.05</b>, now we can check the effect size and statistical power of our experiment in order to safely reject the null hypothesis.

```python
def Cohen_d(group1, group2):
    diff = group1.mean() - group2.mean()
    n1, n2 = len(group1), len(group2)
    var1 = group1.var()
    var2 = group2.var()
    pooled_var = (n1 * var1 + n2 * var2) / (n1 + n2)
    d = diff / np.sqrt(pooled_var)
    return d
```

```python
d_dsc = Cohen_d(df_no_dsc["Quantity"], df_dsc["Quantity"])
d_dsc
```

With the resulting effect size we can say that discount has a small effect on the quantities in an order.

```python
# Checking Statistical power.
power_dsc = TTestIndPower()
```

```python
power_dsc.solve_power(effect_size= d_dsc, nobs1= len(df_no_dsc["Quantity"]), alpha= 0.05)
```

With a Statistical Power of 1, we can safely say that there is enough evidence to reject the null hypothesis and accept that discounts have a significant effect on the quantities of a product in an order. Now we will look at the different discount levels to determine which has a greater impact on order quantities. In order to do this I will run an ANOVA test followed by Tukey's Test which will show the differences in between each of the groups. 

```python
# I will look at each discount level and then create a dataframe for each different level.
df_dsc["Discount"].value_counts()
```

```python
# From the result above I will drop the .03, .04, .06, .01 values because they do not have sufficient observations for our
# statistical test. Using these different data sets I will create a DataFrame that contains no discounts, discounts and the
# quantities from each order.
dsc_0 = df_quan_disc[df_quan_disc["Discount"] == 0].reset_index(drop = True)
dsc_5 = df_dsc[df_dsc["Discount"] == 0.05].reset_index(drop = True)
dsc_10 = df_dsc[df_dsc["Discount"] == 0.10].reset_index(drop = True)
dsc_15 = df_dsc[df_dsc["Discount"] == 0.15].reset_index(drop = True)
dsc_20 = df_dsc[df_dsc["Discount"] == 0.20].reset_index(drop = True)
dsc_25 = df_dsc[df_dsc["Discount"] == 0.25].reset_index(drop = True)
```

```python
df_list = [dsc_0, dsc_5, dsc_10, dsc_15, dsc_20, dsc_25]
tot_quan = []
for quan in df_list:
    tot_quan += list(quan["Quantity"])
len(tot_quan)
```

```python
df_list2 = [dsc_0, dsc_5, dsc_10, dsc_15, dsc_20, dsc_25]
dsc_levels = []
for level in df_list2:
    dsc_levels += list(level["Discount"])
len(dsc_levels)
```

```python
tukey_df = pd.DataFrame({"Discount" : dsc_levels, "Quantity" : tot_quan})
tukey_df.head(20)
```

```python
# ANOVA test
formula = "Quantity ~ C(Discount)"
lm = ols(formula, tukey_df).fit()
table = sm.stats.anova_lm(lm, typ = 2)
print(table)
```

```python
from statsmodels.stats.multicomp import pairwise_tukeyhsd
```

```python
multi_com = pairwise_tukeyhsd(tukey_df["Quantity"], tukey_df["Discount"], alpha= 0.05)
print(multi_com)
```

According to the test above, we can see the comparisons in between the group with <b>0.0</b> discount and the <b>0.05</b>, <b>0.10</b>, <b>0.15</b>, <b>0.20</b> and <b>0.25</b> discount levels. The reject column with True values tells us that we can reject the null hypothesis for 5%, 15%, 20% and 25% discount levels.  

```python
# Test the effect of each of the discount levels to the quantity ordered of a product
es_5 = Cohen_d(dsc_5["Quantity"], dsc_0["Quantity"])
print("Effect Size for 5% discount: ", es_5)
es_10 = Cohen_d(dsc_10["Quantity"], dsc_0["Quantity"])
print("Effect Size for 10% discount: ", es_10)
es_15 = Cohen_d(dsc_15["Quantity"], dsc_0["Quantity"])
print("Effect Size for 15% discount: ", es_15)
es_20 = Cohen_d(dsc_20["Quantity"], dsc_0["Quantity"])
print("Effect Size for 20% discount: ", es_20)
es_25 = Cohen_d(dsc_25["Quantity"], dsc_0["Quantity"])
print("Effect Size for 25% discount: ", es_25)
```

All but the 10% discount level have about a medium effect on the quantity of a product in an order.

```python
# Test for power
pow_5 = power_dsc.solve_power(effect_size = es_5, nobs1 = len(dsc_0["Quantity"]), alpha= 0.05)
print("Power Statistic for 5% discount level: ", pow_5)
pow_10 = power_dsc.solve_power(effect_size = es_10, nobs1 = len(dsc_0["Quantity"]), alpha= 0.05)
print("Power Statistic for 10% discount level: ", pow_10)
pow_15 = power_dsc.solve_power(effect_size = es_15, nobs1 = len(dsc_0["Quantity"]), alpha = 0.05)
print("Power Statistic for 15% discount level: ", pow_15)
pow_20 = power_dsc.solve_power(effect_size = es_20, nobs1 = len(dsc_0["Quantity"]), alpha = 0.05)
print("Power Statistic for 20% discount level: ", pow_20)
pow_25 = power_dsc.solve_power(effect_size = es_25, nobs1 = len(dsc_0["Quantity"]), alpha = 0.05)
print("Power Statistic for 25% discount level: ", pow_25)
```

```python
# Now let's visualize the effect the discount levels have on order quantities
sns.barplot(x = "Discount", y = "Quantity", data = tukey_df, ci = False, palette= "coolwarm")
plt.title("Average Quantity Ordered per Discount Percentage")
plt.show()
```

There is definitely enough evidence to safely reject the null hypothesis and accept that discount has a significant effect on the quantity of a product in an order. And according to our effect sizes, the percentages that have the most effect are the 15% and 25% discounts, just like our bar graph above shows. A recommendation that could be made is to apply this discount to the products that have a hard time selling. 

## 2) Does South America generate more sales than North America?

<b>Null Hypothesis:</b> Average sales are the same or more in South America than in North America.

<b>Alternative Hypothesis:</b> Average sales are less in South America than average sales in North America.

<b>Alpha Level: 0.05</b>

```python
# Query for data
cur.execute("""SELECT c.Region, od.OrderId, od.Quantity, od.UnitPrice, od.Discount
               FROM Customer c
               JOIN "Order" o 
               ON c.Id = o.CustomerId
               JOIN OrderDetail od
               ON od.OrderId = o.Id""")
NA_SA_df = pd.DataFrame(cur.fetchall())
NA_SA_df.columns = [x[0] for x in cur.description]
NA_SA_df.head(20)
```

Now we have the data by region and the Quantity, UnitPrice, and Discount will help in calculating the sales per order. Now I separate the orders from North and South America. 

```python
SA_df = NA_SA_df[NA_SA_df["Region"] == "South America"]
SA_df.head(20)
```

```python
NA_df = NA_SA_df[NA_SA_df["Region"] == "North America"]
NA_df.head(20)
```

```python
# For each dataframe I will create a new column that will contain the total cost per order.
SA_df["Cost_per_Order"] = SA_df["UnitPrice"] * SA_df["Quantity"] * (1 - SA_df["Discount"])
SA_df.head()
```

```python
NA_df["Cost_per_Order"] = NA_df["UnitPrice"] * NA_df["Quantity"] * (1 - NA_df["Discount"])
NA_df.head()
```

```python
# There is no need for the Quantity, UnitPrice or Discount columns
SA_df.drop(["Quantity", "UnitPrice", "Discount"], axis = 1, inplace = True)
NA_df.drop(["Quantity", "UnitPrice", "Discount"], axis = 1, inplace = True)
```

```python
SA_df.head()
```

```python
NA_df.head()
```

```python
SA_df = SA_df.groupby(["OrderId", "Region"])["Cost_per_Order"].sum().reset_index()
SA_df.head()
```

```python
NA_df = NA_df.groupby(["OrderId", "Region"])["Cost_per_Order"].sum().reset_index()
NA_df.head()
```

```python
# Lets see how these two regions compare
# fig, (ax1, ax2) = plt.subplots(nrows= 1, ncols= 2, figsize= (10, 10), sharey = True)
# sns.barplot(SA_df["Region"], SA_df["Cost_per_Order"], ci = False, color = "red", ax = ax1)
# sns.barplot(NA_df["Region"], SA_df["Cost_per_Order"], ci = False, color = "blue", ax =ax2)
```

```python
# Lets see the distributions
sns.distplot(SA_df["Cost_per_Order"]);
print("Number of Observations: ", SA_df.shape[0])
```

```python
sns.distplot(NA_df["Cost_per_Order"]);
print("Number of Observations: ", NA_df.shape[0])
```

Seems like the distribution is not totally normal, and with the number of observations over 30, we can use Central Limit Theorem to achieve a more normal distribution.

```python
# Applying Central Limit Theorem
SA_samp = []
for i in range(160):
    samples_SA = SA_df["Cost_per_Order"].sample(n = 100)
    samp_mn_SA = np.mean(samples_SA)
    SA_samp.append(samp_mn_SA)
sns.distplot(SA_samp);
```

```python
NA_samp = []
for i in range(160):
    samples_NA = NA_df["Cost_per_Order"].sample(n = 100)
    samp_mn_NA = np.mean(samples_NA)
    NA_samp.append(samp_mn_NA)
sns.distplot(NA_samp);
```

```python
# Now lets see what difference we can see.
sns.set(rc = {"figure.figsize" : (8, 8)})
sns.set(color_codes = True)
sns.distplot(SA_samp, label = "South America")
sns.distplot(NA_samp, label = "North America")
plt.legend()
plt.xlabel("Cost per Order")
plt.ylabel("")
plt.title("Average Cost per Order")
plt.show()
```

We can see a big difference, now we'll perform a two sample t-test to see if this difference is in fact statistically significant.

```python
results = stats.ttest_ind(SA_samp, NA_samp, equal_var= False)
print("The p-value: ", results[1])
if results[1] < 0.05:
    print("Null Hypothesis rejected due to a p-value lower than our alpha of 0.05")
else:
    print("Failure to reject Null Hypothesis")
```

The Welch's T-test confirms our observation with a p-value well below our <b>alpha</b> of <b>0.05</b>, now we can check the effect size and statistical power of our experiment in order to safely reject the null hypothesis.

```python
SA_NA_d = Cohen_d(SA_df["Cost_per_Order"], NA_df["Cost_per_Order"])
print("Effect Size: ", SA_NA_d)
```

With this result we can say that there is a medium effect or difference between the two groups.

```python
power_dsc.solve_power(effect_size= SA_NA_d, nobs1= len(SA_df["Cost_per_Order"]), alpha= 0.05)
```

Since the power result is above 0.8 , we can safely reject the null hypothesis and conclude that North America generates more sales than South America.

## 4) Do orders that have Dairy Products, generate more revenue than orders without Dairy Products?

<b>Null Hypothesis:</b> Total price for orders without Dairy Products are higher or the same as orders with Dairy Products.

<b>Alternative Hypothesis:</b> Total price for orders with Dairy Products are higher than orders without Dairy Products.

<b>Alpha: 0.05</b>

```python
# Query for data
cur.execute("""SELECT od.OrderId, c.CategoryName, od.Quantity, od.Discount, od.UnitPrice
               FROM OrderDetail od
               JOIN "Order" o
               ON od.OrderId = o.Id
               JOIN Product p
               ON od.ProductId = p.Id
               JOIN Category c
               ON p.CategoryId = c.Id""")
item_cat_df = pd.DataFrame(cur.fetchall())
item_cat_df.columns = [x[0] for x in cur.description]
item_cat_df.head()
```

```python
# Explore the data
item_cat_df2 = item_cat_df.copy()
item_cat_df2["Total_Price_per_Order"] = item_cat_df2["Quantity"] * item_cat_df2["UnitPrice"] * (1 - item_cat_df2["Discount"])
item_cat_df2.head()
```

```python
plt.figure(figsize = (16, 7))
sns.barplot(item_cat_df2["CategoryName"], item_cat_df2["Total_Price_per_Order"], ci = False, palette = custom_palette)
plt.title("Total Price per Order by Category")
plt.show()
```

Dairy products seem to be a significant part of every order. Not as significant as Meat/Poultry, but that is probably due to higher unit prices. 

```python
# Now I will pull the orders that are from the Dairy Products category then procede to create two dataframes;
# one for the orders without Dairy Products and the orders with Dairy Products, along with the total price per order. 
without_dairy = item_cat_df.loc[item_cat_df["CategoryName"] != "Dairy Products"]
with_dairy = item_cat_df.loc[item_cat_df["CategoryName"] == "Dairy Products"]
```

```python
# Grouping together the data needed.
without_dairy = without_dairy.groupby(["OrderId", "CategoryName"])["Total_Price"].sum().reset_index()
without_dairy.head()
```

```python
with_dairy = with_dairy.groupby(["OrderId", "CategoryName"])["Total_Price"].sum().reset_index()
with_dairy.head()
```

Looks like there is a slight difference in the average total price between the orders. 

```python
# Pull samples for t-test
w_o_dairy_samp = []
for i in range(160):
    samples_w_o = without_dairy["Total_Price"].sample(n = 100)
    samp_mn_w_o = np.mean(samples_w_o)
    w_o_dairy_samp.append(samp_mn_w_o)
sns.distplot(w_o_dairy_samp);
```

```python
w_dairy_samp = []
for i in range(160):
    samples_w = with_dairy["Total_Price"].sample(n = 100)
    samp_mn_w = np.mean(samples_w)
    w_dairy_samp.append(samp_mn_w)
sns.distplot(w_dairy_samp);
```

```python
# T-test 
results = stats.ttest_ind(without_dairy["Total_Price"], with_dairy["Total_Price"], equal_var= False)
print("The p-value: ", results[1])
if results[1] < 0.05:
    print("Null Hypothesis rejected due to a p-value lower than our alpha of 0.05")
else:
    print("Failure to reject Null Hypothesis")
```

```python
# Effect size
dairy_d = Cohen_d(without_dairy["Total_Price"], with_dairy["Total_Price"])
print("Effect Size: ", dairy_d)
```

```python
dairy_pow = power_dsc.solve_power(effect_size = dairy_d, nobs1 = len(without_dairy), alpha = 0.05)
print("Power of t-test: ", dairy_pow)
```

With a power above 0.8, we can safely reject the null hypothesis and say that orders with Dairy Products on average have a higher total price than orders without Dairy, but the effect size reflects a tiny difference in between the two groups of orders.
